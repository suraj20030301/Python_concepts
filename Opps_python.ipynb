```python
class Student:
    def __init__(self):
        self.id=input("Student's ID: ")
        self.student_name=input("Student's full name: ")
        self.age=int(input("Student's Age: "))
        self.marks=int(input("Student's Marks: "))


    def id_data_valid(self):
        if self.age>=20 and self.marks>=0 and self.marks<=100:
            return True
        else:
            return False

    def qualifies_for_addmission(self):
        if self.id_data_valid():
            if self.marks>64:
                return True
            else:
                return False
        else:
            return False
```
